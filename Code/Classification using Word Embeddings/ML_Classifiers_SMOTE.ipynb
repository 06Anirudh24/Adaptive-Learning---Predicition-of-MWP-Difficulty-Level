#ML and DL model initializations

```python
def Decision_tree_classifier (X_train, y_train, X_test, y_test):
  # Create Decision Tree classifer object
  clf = DecisionTreeClassifier()

  # Train Decision Tree Classifer
  clf = clf.fit(X_train,y_train)

  #Predict the response for test dataset
  y_pred = clf.predict(X_test)

  print("Decision Tree Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def GNB_classifier (X_train, y_train, X_test, y_test):
  #Create a Gaussian Classifier
  gnb = GaussianNB()

  #Train the model using the training sets
  gnb.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = gnb.predict(X_test)

  print("Gaussian Naive Bayes Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def KNN_classifier (X_train, y_train, X_test, y_test):
  #Create KNN Classifier
  knn = KNeighborsClassifier(n_neighbors=5)

  #Train the model using the training sets
  knn.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = knn.predict(X_test)

  print("KNN Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def svm_classifier (X_train, y_train, X_test, y_test):
  #Create a Support Vector Machine Classifier
  clf = svm.SVC(kernel='linear') # Linear Kernel

  #Train the model using the training sets
  clf.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = clf.predict(X_test)

  print("SVM Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def random_forest_classifier (X_train, y_train, X_test, y_test):
  #Create a Gaussian Classifier
  clf=RandomForestClassifier(n_estimators=100)

  #Train the model using the training sets y_pred=clf.predict(X_test)
  clf.fit(X_train,y_train)

  #Predict the response for test dataset
  y_pred=clf.predict(X_test)
  print("RF Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def ada_boost_classifier (X_train, y_train, X_test, y_test):
  # Create adaboost classifer object
  abc = AdaBoostClassifier(n_estimators=50,
                          learning_rate=1)
  # Train Adaboost Classifer
  model = abc.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = model.predict(X_test)

  print("Ada-Boost Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def gradient_boost_classifier (X_train, y_train, X_test, y_test):
  # Create adaboost classifer object
  gbc = GradientBoostingClassifier(n_estimators=50,
                          learning_rate=1)
  # Train Adaboost Classifer
  model = gbc.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = model.predict(X_test)

  print("Gradient-Boost Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
def perceptron_classifier (X_train, y_train, X_test, y_test):
  p = Perceptron(random_state=150)
  p.fit(X_train, y_train)
  predictions_train = p.predict(X_train)
  predictions_test = p.predict(X_test)
  train_score = accuracy_score(predictions_train, y_train)
  #print("score on train data: ", train_score)
  test_score = accuracy_score(predictions_test, y_test)
  print("Perceptron Accuracy:", test_score)
```

#Data Loading, Pre-prcossing, and Classification

```python
path = "/content/drive/MyDrive/PTV Sir Project/BERT_algebra_with_levels.csv" 
df_level = pd.read_csv(path)
df_level = df_level["Level"]
df_level.head()
```

```python
print("ANOVA Accuracy Scores:")

for i in range(1, 26):

  path = "/content/drive/MyDrive/ML_Classifiers_Data/ANOVA Features/"+str(i)+".csv" 
  df_data = pd.read_csv(path)
  df_data = df_data.iloc[:, 1:] #Drop 1st column

  #Concatenate data df with levels df
  df = pd.concat([df_data, df_level], axis=1)
  df.head()

  #Data Shuffling
  df = df.sample(frac = 1) #frac denotes the fraction of samples to be returned. 

  #Build 3 levels
  df.loc[df.Level == 2, "Level"] = 1
  df.loc[df.Level == 4, "Level"] = 3
  df["Level"]

  #train-test split
  trdata = df.sample(frac=0.7, random_state=25) #Training data = 70%
  tsdata = df.drop(trdata.index) #Testing data = 30%
  X_train = trdata.iloc[:,:-1]
  y_train = trdata.iloc[:,-1]
  X_test = tsdata.iloc[:,:-1]
  y_test = tsdata.iloc[:,-1]

  #SMOTE
  sm = SMOTE()
  X_train, y_train = sm.fit_resample(X_train, y_train.ravel())

  print("--------------------------------------CSV File Number {}--------------------------------------".format(i))

  #Model Calling
  Decision_tree_classifier(X_train, y_train, X_test, y_test)
  GNB_classifier(X_train, y_train, X_test, y_test)
  KNN_classifier(X_train, y_train, X_test, y_test) 
  svm_classifier(X_train, y_train, X_test, y_test)
  random_forest_classifier(X_train, y_train, X_test, y_test)
  ada_boost_classifier(X_train, y_train, X_test, y_test)
  gradient_boost_classifier(X_train, y_train, X_test, y_test)
  perceptron_classifier(X_train, y_train, X_test, y_test)
```
